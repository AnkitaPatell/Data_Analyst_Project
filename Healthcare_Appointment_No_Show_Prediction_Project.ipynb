```python
# Check how many duplicates you have
print(f"Number of duplicate rows: {df.duplicated().sum()}")

# Remove duplicate rows (keep the first occurrence)
df = df.drop_duplicates()
```

```python
X = df.drop(columns=['No-show'])   # Features
y = df['No-show']                  # Target (already encoded as 0/1)

```

**Splitting into train and test set**

```python
from sklearn.model_selection import train_test_split

# Split into 80% train and 20% test
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,        # 20% test size
    random_state=42,      # for reproducibility
             # keeps the same class distribution in train/test
)
```

**Training Model**

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Create the Decision Tree classifier (you can set parameters here)
#clf = DecisionTreeClassifier(max_depth=5, random_state=42) / Test Accuracy: 0.7962
clf = DecisionTreeClassifier(
    max_depth=7,
    min_samples_split=10,
    min_samples_leaf=5,
    criterion='entropy',
    random_state=42
)

# Train the model
clf.fit(X_train, y_train)
```

```python
?DecisionTreeClassifier
```

```python
# Predict on test data
y_pred = clf.predict(X_test)

```

```python
y_pred
```

```python
submission = pd.DataFrame({"id": range(0,X_test.shape[0]), "pollution_value": y_pred})
submission.to_csv('submission.csv',index=False)
```

```python
# Evaluate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Test Accuracy: {accuracy:.4f}")
```

**Testing model**

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Test Accuracy: {accuracy:.4f}")

# Detailed metrics
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Confusion matrix
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```
